```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

IMPORTING DATASET

```python
data=pd.read_csv('http://bit.ly/w-data')
data.info()
```

```python
data.shape
```

```python
data.duplicated().sum()
```

```python
data.describe()
```

```python
data.plot(x='Hours',y='Scores',style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
X=data.iloc[:, :-1].values
Y=data.iloc[:, 1].values
trainx ,testx,trainy,testy=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
reg=LinearRegression()
reg.fit(trainx,trainy)
pred=reg.predict(testx)
```

```python
df=pd.DataFrame({'Actual':testy,'Predicted':pred})
df
```

ACCURACY

```python
metrics.r2_score(testy,pred)
```

```python
print("Mean Squared Error" , metrics.mean_squared_error(testy,pred))
print("Mean Absolute Error", mean_absolute_error(testy,pred))
```

### *PREDICTED SCORE OF A STUDENT STUDYING 9.25 HRS/DAY*

```python
hours=[9.25]
result=reg.predict([hours])
print("Marks={}" ,format(round(result[0],8)))
```
